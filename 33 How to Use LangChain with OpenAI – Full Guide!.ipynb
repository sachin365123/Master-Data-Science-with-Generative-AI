```python
from langchain_openai.chat_models import ChatOpenAI
```

```python
chat = ChatOpenAI(model_name = 'gpt-4', 
                  model_kwargs = {'seed':365}, 
                  temperature = 0, #Controls the level of randomness in responses, with a lower number corresponding to more deterministic answers
                  max_tokens = 100)
```

```python
response = chat.invoke(''' I'm planning to switch IT company. Could you suggest some developer role names? ''')
```

```python
print(response.content)
```

# System and Human Messages

```python
from langchain_core.messages import SystemMessage, HumanMessage
```

```python
chat = ChatOpenAI(model_name = 'gpt-4', 
                  model_kwargs = {'seed':365},
                  temperature = 0,
                  max_tokens = 100)
```

```python
message_s = SystemMessage(content = ''' You are Marv, a chatbot that reluctantly answers
questions with sarcastic responses. ''')
message_h = HumanMessage(content = ''' I'm planning to switch IT company. Could you suggest some developer role names? ''')
```

```python
response = chat.invoke([message_s, message_h])
```

```python
print(response.content)
```

# AI Messages

```python
from langchain_core.messages import HumanMessage, AIMessage
```

```python
message_h_it = HumanMessage(content = ''' I'm planning to switch IT company. Could you suggest some developer role names? ''')
# AI Message: Assistant-role message in Open AI's API, 
# The response variables are AI messages, are used in few shot Prompting messages
# Few shot Prompting: We give the model a set of example responses to learn from. 
# so that it later generates responses that align better with our expectations.
message_ai_it = AIMessage(content = ''' Oh, absolutely. Because nothing screams "I'm a responsible recruitment team" 
like asking a chatbot to name your new it jobs. How about "Prompt engineer" (if it's a literary Agent AI)? ''')

message_h_nonit = HumanMessage(content = ''' I've recently adopted a non-IT field. Can you suggest some non-IT developer role names? ''')
message_ai_nonit = AIMessage(content = ''' Oh, absolutely. Because nothing screams "I'm a unique and creative recruitment team member" 
like asking a chatbot to name your non-IT developer role. How about "Data Entry Person", "Hardware Repair Man", or "Mobile Opererator"? ''')

message_h_bank = HumanMessage(content = ''' I've recently adopted a Commerce field. Can you suggest some Commerce related role names? ''')
```

```python
# Organize the list with three items in this order.

# The initial inquiry about IT company developer role names, followed by the AI's example response and concluding with another

# query regarding non-IT field job names.

response = chat.invoke([message_h_it, message_ai_it, message_h_nonit])
```

```python
print(response.content)
```

```python
# Organize the list with three items in this order.

# The initial inquiry about non-IT field job names, followed by the AI's example response and concluding with another

# query regarding non-IT field job names.

response = chat.invoke([message_h_it, message_ai_it, message_h_nonit, message_ai_nonit, message_h_bank])
```

```python
print(response.content)
```
